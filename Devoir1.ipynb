Exercice 7: 

Revoir l’exercice 5, en utilisant les listes et les fonctions pour saisir des nombres réels et pour calculer les fonctions suivantes (sans utiliser Numpy) :
la somme
la moyenne
le produit
le max
le min
l’écart-type
la variance
la covariance
Pour effectuer cet exercice, vous aurez besoin d'utiliser la fonction sqrt pour l'utilisation de la fonction racine carée
Pour cela, ajoutez en haut de l'exercice la ligne suivante :
from math import sqrt

```python
from math import sqrt

def saisir_nombres():
    nombres = input("Entrez des nombres réels séparés par des espaces : ")
    return list(map(float, nombres.split()))

def somme(liste):
    total = 0
    for nombre in liste:
        total += nombre
    return total

def moyenne(liste):
    return somme(liste) / len(liste)

def produit(liste):
    prod = 1
    for nombre in liste:
        prod *= nombre
    return prod

def max_val(liste):
    max_value = liste[0]
    for nombre in liste:
        if nombre > max_value:
            max_value = nombre
    return max_value

def min_val(liste):
    min_value = liste[0]
    for nombre in liste:
        if nombre < min_value:
            min_value = nombre
    return min_value

def ecart_type(liste):
    moy = moyenne(liste)
    var = variance(liste, moy)
    return sqrt(var)

def variance(liste, moy=None):
    if moy is None:
        moy= moyenne(liste)
    var = 0
    for nombre in liste:
        var += (nombre - moy) ** 2
    return var / len(liste)

def covariance(liste_x, liste_y):
    moy_x = moyenne(liste_x)
    moy_y = moyenne(liste_y)
    cov = 0
    for x, y in zip(liste_x, liste_y):
        cov += (x - moy_x) * (y - moy_y)
    return cov / len(liste_x)


nombres = saisir_nombres()
print("Somme :", somme(nombres))
print("Moyenne :", moyenne(nombres))
print("Produit :", produit(nombres))
print("Max :", max_val(nombres))
print("Min :", min_val(nombres))
print("Écart-type :", ecart_type(nombres))
print("Variance :", variance(nombres))

print("Pour la covariance, entrez deux listes de même longueur.")
nombres_x = saisir_nombres()
nombres_y = saisir_nombres()
print("Covariance :", covariance(nombres_x, nombres_y))
```

```python
def personne():
  personnes = [] 
  while True:
    
      nom = input("Quel est votre nom ? ")
      prenom = input("Quel est votre prénom ? ")
      genre = input("Quel est votre genre ? ")
      age = int(input("Quel est votre âge ? "))
      personnes.append((prenom, nom, genre, age))
      continuer = input("Voulez-vous ajouter une autre personne ? (oui/non) : ")
      if continuer.lower() != 'oui':
         break  

  return personnes

def lister_personnes(liste_personnes):
  print("\nListe des personnes :")
  for i, (prenom, nom, genre, age) in enumerate(liste_personnes, start=1):
      print(f"{i}. {prenom} {nom}, {genre}, {age} ans")

def chercher_personne(liste_personnes, nom):
  for prenom, nom_personne, genre, age in liste_personnes:
      if nom_personne.lower() == nom.lower():
          return (prenom, nom_personne, genre, age)  
  return None  


liste_de_personnes = personne()  

lister_personnes(liste_de_personnes)

nom_a_chercher = input("\nEntrez le nom de la personne à chercher : ")
resultat = chercher_personne(liste_de_personnes, nom_a_chercher)

if resultat:
  print(f"Personne trouvée : {resultat[0]} {resultat[1]}, {resultat[2]}, {resultat[3]} ans")
else:
  print("Personne non trouvée.")
```
